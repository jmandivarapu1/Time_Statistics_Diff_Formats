### Will show the time required to read from and write to diffrent data formats in Python

### Create a dataframe of Size 50K records and having 200 Columns
df = pd.DataFrame(np.random.randint(1,100000,size=(50000, 200)))


```python
import sqlite3
import os
from pandas.io import sql
import random
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randint(1,100000,size=(50000, 200)))
print(df.shape)
def test_sql_write(df):
    if os.path.exists('test.sql'):
        os.remove('test.sql')
    sql_db = sqlite3.connect('test.sql')
    df.to_sql(name='test_table', con=sql_db)
    sql_db.close()

def test_sql_read():
    sql_db = sqlite3.connect('test.sql')
    pd.read_sql_query("select * from test_table", sql_db)
    sql_db.close()

def test_hdf_fixed_write(df):
    df.to_hdf('test_fixed.hdf','test',mode='w')
    print(os.path.getsize('test_fixed.hdf'))

def test_hdf_fixed_read():
    pd.read_hdf('test_fixed.hdf','test')
    

def test_hdf_fixed_write_compress(df):
    df.to_hdf('test_fixed_compress.hdf','test',mode='w',complib='blosc')
    print(os.path.getsize('test_fixed_compress.hdf'))

def test_hdf_fixed_read_compress():
    df1=pd.read_hdf('test_fixed_compress.hdf','test')
    print(df1.shape)

def test_hdf_table_write(df):
    df.to_hdf('test_table.hdf','test',mode='w',format='table')
    print(os.path.getsize('test_table.hdf'))

def test_hdf_table_read():
    pd.read_hdf('test_table.hdf','test')

def test_hdf_table_write_compress(df):
    df.to_hdf('test_table_compress.hdf','test',mode='w',complib='blosc',format='table')
    print(os.path.getsize('test_table_compress.hdf'))

def test_hdf_table_read_compress():
    pd.read_hdf('test_table_compress.hdf','test')

def test_csv_write(df):
    df.to_csv('test.csv',mode='w')
    print(os.path.getsize('test.csv'))

def test_csv_read():
    pd.read_csv('test.csv',index_col=0)
df
```

### Now we will write the entire dataframe into diffrent formats and check the write time and file sizes

```python
%%time 
test_sql_write(df)
```

```python
%%time 
test_hdf_fixed_write(df)
```

```python
%%time
test_hdf_table_write(df)
```

### As you can see that the below fixed write compress will give you better time and less disk storage as the ouput file size is only 25MB

```python
%%time
test_hdf_fixed_write_compress(df)
```

```python
%%time
test_hdf_table_write_compress(df)
```

```python
%%time
test_csv_write(df)
```
